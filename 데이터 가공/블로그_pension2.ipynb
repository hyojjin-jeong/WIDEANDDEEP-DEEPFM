csv 파일 list로 변환

```python
place_name = place['place_name'].tolist()
detail_list = couple_vlog['블로그 내용'].tolist()
```

```python
place_name.remove('공감')
place_name.remove('강릉시')
```

# Okt

```python
from konlpy.tag import Okt
okt = Okt()
```

# 커플 여행 블로그에서 여행경로 탐색하기

```python
p = []
re = []
couple = []
s = set(place_name)
for i in range(len(detail_list)):
  data = detail_list[i]
  p.clear()
  p = okt.phrases(detail_list[i])
  re.append(p)
  ree = [x for x in re[i] if x in s]
  couple.append(ree)
  print(i,ree)
couple = [v for v in couple if v]
```

```python
couple
```

# pandas data 프레임

블로그 여행경로와 장소 정보 합치기

```python
r2 = []
no_p = p_li.copy()
for i in range(len(couple)):
  for j in range(len(couple[i])):
    a = couple[i][j]
    y = list(zip(*p_li))[1].index(a)
    r1 = [i+6207,'pension',j+1] + p_li[y]
    r2.append(r1)
    if p_li[y] in list(zip(*no_p))[1]:
      no_p.remove(p_li[y])
#print(r2)   
```

```python
r2
```

```python
total = pd.DataFrame(r2)
total.columns = ['user_id','trip','course','place_id','place_name','place_category','rating','review','place_address','place_lat','place_lng']
total['y'] = 1
```

```python
total
```

여행자가 가지 않은 장소와 여행자 합치기

```python
from random import *
r3 = []
for i in range(len(couple)):
  for j in range(len(no_p)):
    k = randint(1,10)
    r4 = [i+6207,'pension',k] + no_p[j]
    r3.append(r4)
```

```python
total1 = pd.DataFrame(r3)
total1.columns = ['user_id','trip','course','place_id','place_name','place_category','rating','review','place_address','place_lat','place_lng']
total1['y'] = 0
```

```python
total1
```

```python
result1 = pd.concat([total,total1], ignore_index=True)
result1
```

```python
result1.to_csv('/content/drive/MyDrive/졸업프로젝트/강원도 명소 데이터/final data/pension2_full.csv', index = False, encoding='utf-8')
```
