```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

# 강원도 춘천 명소 크롤링하기

사이트 주소

```python
driver.get("https://map.kakao.com/")
```

검색어 입력

```python
searchbox = driver.find_element_by_id("search.keyword.query")
searchbox.send_keys("강원도 강릉 여행")
```

검색 버튼 클릭

```python
driver.find_element_by_id("search.keyword.submit").send_keys(Keys.ENTER)
time.sleep(1)
```

장소 더보기 클릭

```python
driver.find_element_by_css_selector('#info\.main\.options > li.option1 > a').send_keys(Keys.ENTER)
time.sleep(1)
```

크롤링 시작

```python
pp = []
result = []
for idx in range(1,40):

  place_lists = driver.find_elements_by_css_selector('.placelist > .PlaceItem')

  for place in place_lists:
    place_name = place.find_element_by_css_selector('div.head_item.clickArea > strong > a.link_name').text
    place_cat = place.find_element_by_css_selector('div.head_item.clickArea > span').text

    try:
      place_rate = place.find_element_by_css_selector('.rating > .score > .num').text
    except:
      place_rate = '0.0'

    try:
      place_review = place.find_element_by_css_selector('.rating > .review > em').text
    except:
      place_review = '0'

    place_add = place.find_element_by_css_selector('div.info_item > div.addr > p:nth-child(1)').text

    pp = [place_name,place_cat,place_rate,place_review,place_add]
    result.append(pp)

  try:  
    if (idx % 5) == 0:
      driver.find_element_by_css_selector('#info\.search\.page\.next').send_keys(Keys.ENTER)
      time.sleep(1)
    else:
      driver.find_element_by_css_selector('#info\.search\.page\.no' + str((idx % 5) + 1)).send_keys(Keys.ENTER)
      time.sleep(1)
  except:
    break
```

```python
dataframe = pd.DataFrame(result)
dataframe.columns = ['place_name','category','rating','review','address']
```
