```python
import googlemaps
import pandas as pd
```

```python
place = pd.read_csv('/content/drive/MyDrive/졸업프로젝트/강원도 명소 데이터/강릉_춘천_통합_최종.csv', encoding='utf-8')
```

```python
place = place.dropna(axis=0)
```

```python
place = place.reset_index()
```

```python
places = place['address'].tolist()
```

```python
len(places)
```

```python
my_key = "AIzaSyCqJKGNfnUIRK1E9rcCNFQdf4P8bj5Byac"
maps = googlemaps.Client(key=my_key)  # my key값 입력
lat = []  #위도
lng = []  #경도

i=0
for place in places:   
    i = i + 1
    try:
        geo_location = maps.geocode(place)[0].get('geometry')
        lat.append(geo_location['location']['lat'])
        lng.append(geo_location['location']['lng'])
        

    except:
        lat.append('')
        lng.append('')
        print("%d번 인덱스 위치를 찾는데 실패했습니다."%(i))
```

```python
df1 = pd.DataFrame(lat, columns=['place_lat'])
df2 = pd.DataFrame(lng, columns=['place_lng'])
```

```python
total = pd.concat([place, df1, df2], axis=1, ignore_index=True)
```

```python
total = total.drop(0, axis=1)
```

```python
total.columns = ['place_id', 'place_name', 'category', 'rating', 'review', 'place_address', 'place_lat', 'place_lng']
```
